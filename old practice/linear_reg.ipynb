```python
import numpy as np

x_train= np.array([1.,2.])
y_train= np.array([300.,500.])

    # size of input arrray in m //
m=x_train.shape[0]
print(m)

```

```python
import matplotlib.pyplot as plt


plt.title("Housing Prices")
plt.xlabel('Size (1000 sqft)')
plt.ylabel('Price in 1000 dollars')
plt.scatter(x_train,y_train,marker='x',c='r')

plt.show()
```

```python
w=100
b=200

# making a model fuction to train machine

def compute_model_output(x,w,b):
    m=x.shape[0]
    f_wb= np.zeros(m)
    for i in range(m):
        f_wb[i]=(w*x[i])+b
    return f_wb
```

```python
temp_fwb= compute_model_output(x_train,w,b)

plt.plot(x_train,temp_fwb,c='b',label='Our Prediction')

plt.scatter(x_train,y_train,marker='x',c='r',label='Actual values')
plt.title("Housing Prices")
plt.xlabel('Size (1000 sqft)')
plt.ylabel('Price in 1000 dollars')
plt.legend()
plt.show()
```
